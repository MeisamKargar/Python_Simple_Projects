```python
try:
    Inp = input('Enter your numbers: ')
    Inp = Inp.split()
    for i in Inp:
        Operator1 = int(Inp[0])
        Operator2 = int(Inp[2])
    if Inp[1] == '+':
        res = Operator1 + Operator2
        print(f'{Operator1} + {Operator2} = {res}')
    elif Inp[1] == '-':
        res = Operator1 - Operator2
        print(f'{Operator1} - {Operator2} = {res}')
    elif Inp[1] == '*':
        res = Operator1 * Operator2
        print(f'{Operator1} * {Operator2} = {res}')
    elif Inp[1] == '/':
        res = Operator1 / Operator2
        print(f'{Operator1} / {Operator2} = {res}') 
except:
    print('you enter 0 number and you have a zero devision error')
```

```python
def move_files(list_of_files): 
    try:
        for i in list_of_files:# i = test.py ===> folder_py/
            ext = i.split(".") # ext = ["test", "py"]
            ext = ext[-1].lower() # ext = "py"
            os.rename(i, f"folder_{ext}/{i}") # folder_py/test.py
    except FileNotFoundError:
        print('FileNotFound')
```

<h5 style="background-color:#1765e3;text-align:left;border-radius:40px;border: 1px;padding: 10px;color:white"><b>1- Write a program that reads all the txt files on the desktop and creates a txt file as follows (300)</h5>

##### In the first column, the path and name of each file, in the second column, specify the number of vowels in the file, and in the following columns, specify the number for each vowel.

```python
import os

desktop_path = os.path.expanduser("C:/Users/meisam/Desktop/Newfolder")
output_file = os.path.join(desktop_path, "vowel_counts.txt")

file_list = [f for f in os.listdir(desktop_path) if f.endswith(".txt")]

with open(output_file, "w") as output:
    for file_name in file_list:
        file_path = os.path.join(desktop_path, file_name)
        with open(file_path, "r") as file:
            file_contents = file.read()

            vowel_counts = {"a": 0, "e": 0, "i": 0, "o": 0, "u": 0}

            for char in file_contents:
                char = char.lower()

                if char in vowel_counts:
                    vowel_counts[char] += 1

            output.write(f"File: {file_name}\n")

            output.write("Vowel Counts: ")
            for vowel, count in vowel_counts.items():
                output.write(f"{vowel}: {count}, ")
            output.write("\n---\n")

print(f"Vowel counts file created: {output_file}")
```
